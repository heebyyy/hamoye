```python
dataset = df.drop(["date", "lights"], axis = 1)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(dataset), columns = dataset.columns)
```

```python
scaled_data.head()
```

```python
que_12 = scaled_data[["T2", "T6"]]
que_12.head()
```

```python
X = que_12[["T2"]]
y = que_12[["T6"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin_mod = LinearRegression()
```

```python
lin_mod.fit(X_train, y_train)
lin_pred = lin_mod.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, lin_pred)
r2
```

```python
import numpy as np
```

```python
que_13 = mean_absolute_error(y_test, lin_pred)
round(que_13, 2)
```

```python
que_14 = np.sum(np.square(y_test - lin_pred))
que_14
```

```python
que_15 = np.sqrt(mean_squared_error(y_test, lin_pred))
```
